def gaussian_filter(size, sigma=1):  
kernel = np.fromfunction(lambda x, y: (1 / (2 \* np.pi \* sigma \*\* 2))
\*  
np.exp(- ((x - (size - 1) / 2) \* 2 + (y - (size - 1) / 2) \* 2) / (2 \*
sigma \*\* 2)), (size, size))  
return kernel / np.sum(kernel)

def apply_filter(image, filter_kernel):  
filtered_image = np.zeros((image.shape\[0\] - filter_kernel.shape\[0\] +
1, image.shape\[1\] - filter_kernel.shape\[1\] + 1))  
for i in range(filtered_image.shape\[0\]):  
for j in range(filtered_image.shape\[1\]):  
filtered_image\[i, j\] = np.sum(image\[i:i+filter_kernel.shape\[0\],  
j:j+filter_kernel.shape\[1\]\] \* filter_kernel)  
return filtered_image
